```python
li_elements = driver.find_elements(By.XPATH, '//*[@id="ReviewsFeed"]/ol/li')
for li_element in li_elements:
    
    date = li_element.find_element(By.CLASS_NAME, 'timestamp_reviewDate__dsF9n').text
    rating = li_element.find_element(By.CLASS_NAME, 'review-rating_ratingLabel__0_Hk9').text
    one_line_review = li_element.find_element(By.CLASS_NAME, 'heading_Heading__BqX5J.heading_Level3__X81KK').text
    job = li_element.find_element(By.CLASS_NAME, 'review-avatar_avatarLabel__P15ey').text
    history = li_element.find_element(By.CLASS_NAME, 'text-with-icon_TextWithIcon__5ZZqT').text
    region = li_element.find_element(By.CLASS_NAME, 'review-avatar_tagsContainer__9NCNs').text.split('\n')[1]
    pros = li_element.find_element(By.CSS_SELECTOR, '[data-test="review-text-PROS"]').text
    cons = li_element.find_element(By.CSS_SELECTOR, '[data-test="review-text-CONS"]').text
    
    show_more_button_class_nm = 'expand-button_ExpandButton__Wevvg'
    show_more_button = li_element.find_element(By.CLASS_NAME, show_more_button_class_nm)
    show_more_button.click()
    time.sleep(2)
    feedback = li_element.find_element(By.CSS_SELECTOR, '[data-test="review-text-FEEDBACK"]').text

    print(date)
    print(rating)
    print(one_line_review)
    print(job)
    print(history)
    print(region)
    print(pros)
    print(cons)
    print(feedback)
    break
```

```python
def apply_filtering(driver):

    """ Filtering Clear """
    
    try:
        filterling_button_class_nm = 'filter-menu_FilterMenuButton__NOhT8'
        filterling_button = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, filterling_button_class_nm))
        )
        filtering_button.click()
        time.sleep(5)
    
        clear_button_class_nm = 'button_Button__MlD2g.button-base_Button__knLaX'
        clear_button = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, clear_button_class_nm))
        )
        actions = ActionChains(driver)
        actions.move_to_element(clear_button).click().perform()
        time.sleep(5)
        
    except Exception as e:
            print(f"Error during button click process: {e}")
```
